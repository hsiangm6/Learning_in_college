```python
import pandas as pd
import numpy as np

```

```python
# 讀取訓練和測試資料gene expression cancer RNA-Seq Data Set-20230503T055527Z-001\gene expression cancer RNA-Seq Data Set
train_label_csv = pd.read_csv("gene expression cancer RNA-Seq Data Set/train_label.csv")
train_data_csv = pd.read_csv("gene expression cancer RNA-Seq Data Set/train_data.csv")
test_label_csv = pd.read_csv("gene expression cancer RNA-Seq Data Set/test_label.csv")
test_data_csv = pd.read_csv("gene expression cancer RNA-Seq Data Set/test_data.csv")
```

```python
print(train_label_csv.shape)
print(train_data_csv.shape)
```

```python
train_label_csv = train_label_csv.drop(columns=['id'])
data = train_data_csv.drop(columns=['id'])
test_label_csv = test_label_csv.drop(columns=['id'])
test_data_csv = test_data_csv.drop(columns=['id'])
train_data_csv=data
```

```python
print(train_label_csv.shape)
print(train_data_csv.shape)
```

```python
from sklearn.preprocessing import OneHotEncoder

# 創建OneHotEncoder對象
enc = OneHotEncoder()

# 將訓練集和測試集的標籤都用於擬合Encoder
enc.fit(train_label_csv)

# 將訓練集和測試集的標籤都轉換成One-Hot編碼
label_encoder = enc.transform(train_label_csv).toarray()
print(label_encoder) #有3類
# 新增一個column並填入0
unknown_col = np.zeros((label_encoder.shape[0], 1))

# 將unknown column插入到最後一欄
label_encoder = np.hstack((label_encoder, unknown_col))

print(label_encoder.shape)
```

```python
pd.DataFrame(label_encoder)
```

```python
# 填補缺失值
data_no_na = np.array(pd.DataFrame(train_data_csv).fillna(pd.DataFrame(train_data_csv).mean()))
data_no_na
```

```python
from sklearn.model_selection import train_test_split

# 分割數據集和標籤為訓練集和測試集
train_data, test_data, train_label, test_label = train_test_split(data_no_na, label_encoder, test_size=0.2, random_state=42)
```

```python
train_data
```

```python
#標準化
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
train_data_scaled = scaler.fit_transform(train_data)
test_data_scaled = scaler.transform(test_data)
```

```python
pd.DataFrame(train_data_scaled)
```

```python
train_data_scaled.shape
```

```python
#第一個參數為 train_data_scaled 和 test_data_scaled，分別代表經過標準化處理後的訓練集和測試集資料。
# 第二個參數中 (train_data_scaled.shape[0], train_data_scaled.shape[1], 1) 的含義為將原本的二維資料轉換成三維資料，其中新的第三維的大小為1。
# 這個新的維度的大小是由 LSTM 模型需要的資料格式決定的，其中第一個維度表示樣本數，第二個維度表示時間步長，第三個維度表示每個時間步中的特徵數
#將原本的二維資料轉換成三維資料，可以讓 LSTM 模型正確地理解資料

train_data_3d = np.reshape(train_data_scaled, (train_data_scaled.shape[0], train_data_scaled.shape[1], 1))
test_data_3d = np.reshape(test_data_scaled, (test_data_scaled.shape[0], test_data_scaled.shape[1], 1))
```

```python
train_data_3d.shape
```

```python
import tensorflow as tf
from tensorflow import keras
```

```python
# 建立模型
from keras.layers import Dropout, BatchNormalization
from keras import regularizers

model = keras.models.Sequential([
    keras.layers.LSTM(48, activation='tanh', return_sequences=False, input_shape=(train_data_3d.shape[1], train_data_3d.shape[2]), kernel_regularizer=regularizers.l2(0.001)),
    Dropout(0.1), #丟棄率為 0.1
    BatchNormalization(), #對上一層輸出進行標準化
    keras.layers.Dense(36, activation='relu', kernel_regularizer=regularizers.l2(0.001)), #輸入維度為 48，輸出維度為 36，激活函數為 relu
    Dropout(0.1),
    BatchNormalization(),
    keras.layers.Dense(24, activation='relu', kernel_regularizer=regularizers.l2(0.001)),
    Dropout(0.1),
    BatchNormalization(),
    keras.layers.Dense(18, activation='relu', kernel_regularizer=regularizers.l2(0.001)),
    Dropout(0.1),
    BatchNormalization(),
    keras.layers.Dense(4, activation='softmax', kernel_regularizer=regularizers.l2(0.001)) #表示預測出的結果屬於 9 個類別中的哪一個
])
#型使用 Adam 優化器，並定義了一個網格搜索需要用到的 learning_rate 參數
optimizer = keras.optimizers.Adam(learning_rate=0.001)
model.summary() #顯示整個模型的結構和參數數量等信息
```

```python
from sklearn.model_selection import KFold

# 定義交叉驗證參數
n_splits = 5 # 設置5折交叉驗證
fold = KFold(n_splits=n_splits, shuffle=True, random_state=42)
#每次選擇其中的k-1個分片作為訓練集，剩下的一個分片作為驗證集。
# 在進行k次交叉驗證時，每個分片都會被當做驗證集一次，因此我們可以得到k個模型，最終結果為這k個模型的平均值。
```

```python
# 增加 callbacks:用於防止過擬合
# 在訓練期間，它會監視指定的度量，例如驗證集的損失或準確度。如果在指定的 epoch 數內，這個度量沒有再改善，則訓練會被停止。
# 在這個例子中，start_from_epoch 參數設置為 150，表示在 150 個 epoch 之後開始監控度量的改善。patience 參數設置為 50，
# 表示如果在接下來的 50 個 epoch 中度量沒有再改善，訓練將被停止。最後，restore_best_weights 參數設置為 True，表示在訓練停止後，
# 模型將被還原為具有最好權重的那個 epoch 的狀態。這樣可以確保模型在最佳狀態下被保存。
early_stopping = keras.callbacks.EarlyStopping(start_from_epoch=150 ,patience=50, restore_best_weights=True)

# 進行交叉驗證
scores = []
for train_index, test_index in fold.split(data):
    # 將數據分為訓練集和測試集
    X_train, X_test = train_data_csv.iloc[train_index], train_data_csv.iloc[test_index]
    y_train, y_test = label_encoder[train_index], label_encoder[test_index]

    # 標準化
    train_data_scaled = scaler.fit_transform(X_train)
    test_data_scaled = scaler.transform(X_test)

    train_data_3d = np.reshape(train_data_scaled, (train_data_scaled.shape[0], train_data_scaled.shape[1], 1))
    test_data_3d = np.reshape(test_data_scaled, (test_data_scaled.shape[0], test_data_scaled.shape[1], 1))

     # 編譯模型
    model.compile(loss='categorical_crossentropy', optimizer=optimizer, metrics=['accuracy'])

    # 訓練模型
    model.fit(train_data_3d, y_train, epochs=500, batch_size=64, validation_data=(test_data_3d, y_test), callbacks=[early_stopping])

    # 評估模型
    score = model.evaluate(test_data_3d, y_test, verbose=0)
    scores.append(score[1])

# 輸出交叉驗證結果
print("Cross-validation accuracy: %.2f%% (+/- %.2f%%)" % (np.mean(scores) * 100, np.std(scores) * 100))
```
